```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('분석완료리뷰.xlsx')
```

```python
df.head()
```

```python
df=df[['영화제목','리뷰','긍정확률','감성2']]
df.head()
```

```python
df['조정확률']=((df['긍정확률']+df['감성2']/100)/2).round(2)
```

```python
df
```

```python
a=df.groupby(['영화제목'])['조정확률'].mean()
```

```python
a
```

```python
b=df[['영화제목','조정확률']]
b.head()
```

```python
b.rename(columns={'영화제목':'영화명'},inplace=True)
b.head()
```

```python
c=b.groupby(['영화명'],as_index=False).mean()
```

```python
c=c.round(2)
```

```python
c
```

```python
result=pd.read_excel('Result.xlsx')
result.head()
```

```python
results=pd.merge(result,c,how='inner')
```

```python
results
```

```python
df=pd.read_excel('분석완료리뷰 진행.xlsx')
df.head()
```

```python
df=df[['영화제목','긍정확률']]
df.rename(columns={'영화제목':'영화명'},inplace=True)
df.head()
```

```python
a=df.groupby(['영화명'],as_index=False).mean()
```

```python
a=a.round(2)
```

```python
a
```

```python
result=pd.read_excel('Result.xlsx')
result.head()
```

```python
results=pd.merge(result,a,how='inner')
results
```

```python
results.to_excel('최종데이터.xlsx',index=False)
```
